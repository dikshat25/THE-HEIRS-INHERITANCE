```python
import numpy as np
from sklearn.neighbors import NearestNeighbors
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.preprocessing import StandardScaler
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

# Replace NaN values in the 'ingredients_name' column with an empty string
df['ingredients_name'] = df['ingredients_name'].fillna("")

# Apply TfidfVectorizer
vectorizer = TfidfVectorizer()
X_ingredients = vectorizer.fit_transform(df['ingredients_name'])

# Check the resulting matrix shape
print(X_ingredients.shape)
```

```python
X_combined=X_ingredients
X_combined
knn=NearestNeighbors(n_neighbors=5,metric='cosine')
knn.fit(X_combined)
```

```python
def recommendation_system(input_features):
  input_ingredients_transformed=vectorizer.transform(input_features)
  distances,indices=knn.kneighbors(input_ingredients_transformed)
  recommended_recipes=df.iloc[indices[0]]
  return recommended_recipes[['name','ingredients_name','image_url']]
```

```python
input_features=['spinach,paneer,butter,salt,cabbage,bread']
recommended_recipes=recommendation_system(input_features)
recommended_recipes
```
